# Implementing Aleksa's Neural Network

```python
def hidden_n_calc(n_input, n_output):
    n_hidden = np.floor(np.add(np.multiply(np.divide(2.0,3.0),n_input),n_output))
    n_hidden2 = np.floor(np.add(np.multiply(np.divide(2.0,3.0),n_hidden),n_output))
    return int(n_hidden), int(n_hidden2)
```

```python
import torch
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class povNet(nn.Module):
    
    def __init__(self, n_input, n_output):
        super(povNet, self).__init__()
        
        n_h1, n_h2 = hidden_n_calc(n_input, n_output)
        self.fc1 = nn.Linear(n_input, n_h1)
        self.fc2 = nn.Linear(n_h1, n_h2)
        self.fc3 = nn.Linear(n_h2, n_output)
        self.dropout = nn.Dropout(0.5)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import StepLR
from torch.utils.data import Dataset, DataLoader
from torch.utils.data import TensorDataset
from torch.utils.data.dataset import random_split
```

```python
# Original CustomDataset class

class CustomDataset(Dataset):
    def __init__(self, data_df, target_df):
        self.data = data_df.values
        self.targets = target_df.values
    def __len__(self):
        return len(self.data)
    def __getitem__(self, index):
        x = torch.tensor(self.data[index], dtype=torch.float32)
        y = torch.tensor(self.targets[index], dtype=torch.float32)
        return x, y
```

```python
import numpy as np

class CustomDataset(Dataset):
    def __init__(self, data, target=None):
        self.data = torch.tensor(data, dtype=torch.float32)
        self.target = torch.tensor(target, dtype=torch.float32) if target is not None else None
        self.length = len(self.data)

    def __len__(self):
        return self.length

    def __getitem__(self, index):
        x = torch.tensor(self.data[index % self.length], dtype=torch.float32)
        y = torch.tensor(self.target[index % self.length], dtype=torch.float32) if self.target is not None else None
        return x, y
```

```python
dataset = CustomDataset(df_train, target_df)
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
net = povNet(df_train.shape[1], 1)

criterion = nn.MSELoss()
optimizer = optim.Adam(net.parameters())

scheduler = StepLR(optimizer, step_size=8, gamma=0.1)

train_acc = list()
test_acc = list()
e_loss = list()
epochs = 41

# Step 4: Train the model
for epoch in range(epochs):
    train_correct = 0
    train_total = 0
    net.train()
    for inputs, labels in train_dataloader:
        # Forward pass
        outputs = net(inputs)

        # Compute loss
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Calculate training accuracy
        predicted = torch.round(outputs)
        train_correct += (predicted == labels).sum().item()
        train_total += labels.size(0)

    train_accuracy = train_correct / train_total

    # Evaluate on the test set
    test_correct = 0
    test_total = 0
    net.eval()
    with torch.no_grad():
        for inputs, labels in test_dataloader:
            # Forward pass
            outputs = net(inputs)

            # Calculate test accuracy
            predicted = torch.round(outputs)
            test_correct += (predicted == labels).sum().item()
            test_total += labels.size(0)

    test_accuracy = test_correct / test_total
    
    scheduler.step()

    if (epoch == 0) or ((epoch+1) % 2 == 0):
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item()}, Train Accuracy: {train_accuracy * 100:.2f}%, Test Accuracy: {test_accuracy * 100:.2f}%")
        #torch.save(net.state_dict(), f'./models/ffnn_at_{epoch}.pth')
    train_acc.append(train_accuracy)
    test_acc.append(test_accuracy)
    e_loss.append(loss.item())
```

```python
# New class for encoding that doesn't need a label column

class encoded_dataset1:

    def __init__(self, image_dir, df, tree, depth=8):
        """ Generate a tree based encoding for a set of images.
        image_dir: location of images
        df: dataframe with information about each image
        tree: the tree used for encoding
        depth: the number of tree levels to be used.
        """
        def bin2int(c):
            ans = 0
            for j in range(len(c)):
                bit = int(c[-(j+1)])
                ans += bit * (2 ** j)
            return ans
  
        self.df = df
        self.rows = df.shape[0]
        self.cols = 2 ** (depth + 1)
        data = zeros([self.rows, self.cols])

        j = 0
        self.codes = []
        for filename, row in df.iterrows():
            filepath = f"{image_dir}/{filename}"
            code = encode_image(filepath, tree)
            self.codes.append(code)
            V = zeros(self.cols)
            for c, a in code:
                V[bin2int(c)] = a

            data[j] = V

            if (j + 1) % 10 == 0:
                print(j, filename, end='\r')
            j += 1

        self.data = data

    def get_df(self):
        return self.df

    def get_slice(self, selection):
        assert selection.shape[0] == self.data.shape[0]
        S = self.data[selection, :]
        return S
```
